```python
def ExtractYear(date):
    date = str(date)
    part = date.split('-')
    shortYear = part[2][0:2]
    if int(shortYear) >= 50:
        return '19'+str(shortYear)
    else:
        return '20'+str(shortYear)
    
df['YEAR'] = [ExtractYear(date) for date in df['DATE']] 
print(df['YEAR'])
```

```python
df[['EVENT_TYPE', 'NAME']] = df[['EVENT_TYPE', 'NAME']].astype('str')
df[['DEATHS', 'INJURIES']] = df[['DEATHS', 'INJURIES']].astype('int64')
df[['BEGIN_LAT','BEGIN_LON']] = df[['BEGIN_LAT','BEGIN_LON']].astype('float')

print(df.dtypes)
```

```python
def ProcessNumeral(value):
    value = str(value)
    if('K' in value or 'k' in value):
        if value == 'K' or value == 'k':
            return 0
        value = value.replace('K','')
        value = value.replace('k','')
        return float(value)*1000
    elif('M' in value or 'm' in value):
        if value == 'M' or value == 'm':
            return 0
        value = value.replace('M','')
        value = value.replace('m','')
        return float(value)*1000000
    elif('B' in value or 'b' in value):
        if value == 'B' or value == 'b':
            return 0
        value = value.replace('B','')
        value = value.replace('b','')
        return float(value)*1000000000
    elif(value == '0'):
        return 0
    else:
        if isinstance(value, float):
            return value
        #print(value)
        return 0

df['DAMAGE_PROPERTY'] = [ProcessNumeral(value) for value in df['DAMAGE_PROPERTY']]
df['DAMAGE_CROPS'] = [ProcessNumeral(value) for value in df['DAMAGE_CROPS']]

df[['DAMAGE_PROPERTY','DAMAGE_CROPS']] = df[['DAMAGE_PROPERTY','DAMAGE_CROPS']].astype('float64')
print(df.dtypes)
```

```python
years = list(range(1950, 2018))
print(years)
cols = ['NAME', 'YEAR', 'EVENT_TYPE', 'DEATHS', 'INJURIES', 'DAMAGE_PROPERTY', 'TEMPERATURE', 'DATE']
Worst_Property_Damage = pd.DataFrame(columns = cols)
for year in years:
    for event in df.EVENT_TYPE.unique():
        dfYear = df.loc[df['YEAR'] == str(year), cols]
        worst = dfYear[dfYear['EVENT_TYPE'] == event].sort_values('DAMAGE_PROPERTY', ascending = False)
        worst = worst.head(10)
        Worst_Property_Damage = Worst_Property_Damage.append(worst)
        
print(Worst_Property_Damage)
Worst_Property_Damage.to_csv(path_or_buf = 'C:\\Users\\Benjamin\\Desktop\\Storm_allyears\\Worst_Property_Damage.csv')
```
